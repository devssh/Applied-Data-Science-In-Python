```python
df = pd.read_excel("BaylorReligionSurveyWaveII(2007).XLSX").fillna(0)
df["DENOM"][df["DENOM"]==0]="ATHEIST"
df["DENOM"]=df["DENOM"].apply(lambda val: "CATHOLIC" if "CATHOLIC" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "PROTESTANT" if "PROTESTANT" in val or "PENTE" in val or
                              "WESL" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "REFORMED" if "REFOR" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "BAPTIST" if "BAPTI" in val or "CBF" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "METHODIST" if "METHODIST" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "LUTHERAN" if "LUTHER" in val or "ELCA" in val or
                              "MISSOU" in val or "LCMS" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "PRESBYTARIAN" if "PRESBYT" in val or "PCUSA" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "ADVENTIST" if "ADVENT" in val or "SOUTHERN" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "APOSTOLIC" if "APOSTO" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "EVANGELICAL" if "EVANG" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "ISLAM" if "SUNNI" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "BUDDHIST" if "NICHIREN" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "GURU" if "GURU" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "JEWISH" if "JEWISH" in val or "JUDA" in val or
                              "JEHOV" in val or "EMMAN" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "FELLOWSHIP" if "FELLOW" in val or "CONGR" in val or 
                              "ASSOC" in val or "FRIEND" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "PAGAN" if "PAGAN" in val or "INDEP" in val or
                              "WICCA" in val or "IDENT" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "UNITARIAN" if "UNITED" in val or "UNIT" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "CONSERVATIVE" if "CONSER" in val or "ORTHO" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "CHRISTIAN" if "ASSEMBL" in val or "CHURCH" in val or 
                              "WORD" in val or "FAITH" in val or val.startswith("ST ") or "CHRIS" in val or
                              "BIBL" in val or "SAINT" in val or "CHAPEL" in val or 
                              "BELIEV" in val or "BORN" in val or "CALV" in val or "GOSPEL" in val or
                              "VINEY" in val or "ANGLICAN" in val or "EPIS" in val or "HOLI" in val or
                              "COVEN" in val or "NAZERENE" in val or "NAZARENE" in val or "ECLECTIC" in val or 
                              "LCA" in val or "INTER" in val or "LDS" in val or "QUAKE" in val or 
                              "GENER" in val or "MORM" in val or "MORI" in val or "OBC" in val else val)
df["DENOM"]=df["DENOM"].apply(lambda val: "AGNOSTIC/ANONYMOUS" if "NON" in val or "GNOSTIC" in val or
                              "ATHEIST" in val or "HORDE" in val else val)
df=df.groupby("DENOM").apply(sum)
df["DENOM"] = df.index

df
```

```python
dfx = df.iloc[[1,3,5,12,13,16,17,18,19]]
dfx = dfx[["WEIGHT","PORN","PREMARSX","HOMOSEX","MARIJUAN","ABORTION","UNDER18","CHILDREN"]]
dfy=dfx["WEIGHT"].copy()
xpos = range(len(dfx.index))
xpos = [pos*7 + 3 for pos in xpos]
xpos1 = [pos-3 for pos in xpos]
xpos2 = [pos-2 for pos in xpos]
xpos3 = [pos-1 for pos in xpos]
xpos4 = [pos for pos in xpos]
xpos5 = [pos+1 for pos in xpos]
xpos6 = [pos+2 for pos in xpos]
xpos7 = [pos+3 for pos in xpos]

gspec = gridspec.GridSpec(3, 1)

plt.figure(figsize=(20,10),dpi=500)
top_histogram = plt.subplot(gspec[0, 1:])
bottom_histogram = plt.subplot(gspec[1:, 1:], sharex=top_histogram)

top_histogram.set_xticks(xpos)
top_histogram.set_xticklabels(dfx.index)

dfx=dfx.apply(lambda num: num/dfx["WEIGHT"])
top_histogram.bar(xpos1,dfx["PORN"], label="Watch Porn", color="black")
top_histogram.bar(xpos2,dfx["PREMARSX"], label="Accept pre-marital sex")
top_histogram.bar(xpos3,dfx["HOMOSEX"], label="Accept Homosexuality")
top_histogram.bar(xpos4,dfx["MARIJUAN"], label="Smoke Marijuana")
top_histogram.bar(xpos5,dfx["ABORTION"], label="Pro Abortion")
top_histogram.bar(xpos6,dfx["UNDER18"], label="Are under 18")
top_histogram.bar(xpos7,dfx["CHILDREN"], label="Have children")

top_histogram.legend(loc="upper left")
top_histogram.set_ylabel("Ratings/ Approvals")
#top_histogram.set_xlabel("Religious Belief")

bottom_histogram.bar(xpos, dfy, width=5)
bottom_histogram.invert_yaxis()
bottom_histogram.get_xaxis().set_visible(False)
bottom_histogram.set_yticks([])
bottom_histogram.set_ylabel("Relative weight in no. of people")

for i,pos in enumerate(xpos):
    bottom_histogram.text(pos-2, dfy.iloc[i]+50, dfy.iloc[i])

plt.show()
```
